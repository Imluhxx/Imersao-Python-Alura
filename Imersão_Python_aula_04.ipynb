```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import mplfinance as mpf
import yfinance as yf        # yf: API que contém dados de ações de bolsas abertas
                             # API: interface que conecta o código a alguma base de dados
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
# Buscar um dado na API
# nome da ação, start = ano que começa, end = ano que termina (datas no fotmato aaaa-mm-dd)

dados = yf.download('PETR4.SA', start= '2023-01-01', end= '2023-12-31')
dados
```

```python
# renomear colunas (renomear todas)

dados.columns = ['Abertura', 'Maximo', 'Minimo', 'Fechamento', 'Fech_Ajust', 'Volume']
dados = dados.rename_axis('Data') # A função rename_axis é utilizada para renomear os rótulos (labels) dos índices de um DataFrame.
dados
```

# Criando o Primeiro Gráfico

```python
# gráfico de fechamento durante o ano todo
# "plotar": desenhar, mostrar uma figura

# Desafio 1: pesquisar o que é uma tupla em python:
# Uma tupla em Python é uma coleção ordenada de elementos, similar a uma lista, porém imutável.
# Isso significa que uma vez que uma tupla é criada, seus elementos não podem ser alterados, adicionados ou removidos.
# As tuplas são definidas utilizando parênteses e podem conter elementos de diferentes tipos de dados. Por exemplo: (1, 'abc', 3.14).
# As tuplas são frequentemente utilizadas para representar coleções de valores relacionados que não devem ser modificados.

dados['Fechamento'].plot(figsize=(10,6)) # (10,6) é uma tupla
plt.title('Variação do Preço por Data', fontsize= 16) #colocar o título e o tamanho da fonte
plt.legend(['Fechamento']) #adicionar a legenda
```

```python
# Começando a criar o gráfico de vela - base

df = dados.head(60).copy()
# Convertendo o índice em uma coluna de data (Criar uma nova coluna com as informações do índice)
df['Data'] = df.index
# Convertendo as datas para o formato numérico do matplotlib
# Isso é necessário para que o Matplotlib possa plotar as datas corretamente no gráfico
df['Data'] = df['Data'].apply(mdates.date2num)  #mdates: método do matplotlib para manipulação de datas
df
```

```python
# Mão na massa!

fig, ax = plt.subplots(figsize= (15, 8)) # tela inicial do gráfico

# Definir a largura dos candles no gráfico
width = 0.7

for i in range(len(df)):  #for: estrutura de repetição / len(df):tamanho do dataframe (60)
  # Determinando a cor do candle
  # Se o preço de fechamento for maior que o de abertura, o candle é verde (a ação valorizou nesse dia).
  # Se for menor, o candle é vermelho (a ação desvalorizou).
  if df['Fechamento'].iloc[i] > df['Abertura'].iloc[i]:
    color = 'green'
  else:
    color = 'red'

  # Desenhando a linha vertical do candle (mecha)
  # Essa linha mostra os preços máximo (topo da linha) e mínimo (base da linha) do dia.
  # Usamos 'ax.plot' para desenhar uma linha vertical.
  # [df['Data'].iloc[i], df['Data'].iloc[i]] define o ponto x da linha (a data), e [df['Mínimo'].iloc[i], df['Máximo'].iloc[i]] define a altura da linha.
  ax.plot([df['Data'].iloc[i], df['Data'].iloc[i]],
              [df['Minimo'].iloc[i], df['Maximo'].iloc[i]],
              color=color,
              linewidth=1)

  #  Definir os retângulos
  ax.add_patch(plt.Rectangle((df['Data'].iloc[i] - width/2, min(df['Abertura'].iloc[i], df['Fechamento'].iloc[i])),
                                width,
                                abs(df['Fechamento'].iloc[i] - df['Abertura'].iloc[i]),
                                facecolor=color))         #add_patch: adicionar objeto. / plt.Rectangle: método do matplotlib

df['MA7'] = df['Fechamento'].rolling(window=7).mean()
df['MA14'] = df['Fechamento'].rolling(window=14).mean()

# Plotando as médias móveis: indica tendência
ax.plot(df['Data'], df['MA7'], color='orange', label='Média Móvel 7 Dias')  # Média de 7 dias
ax.plot(df['Data'], df['MA14'], color='yellow', label='Média Móvel 14 Dias')  # Média de 14 dias
# Adicionando legendas para as médias móveis
ax.legend()

# Formatando o eixo x para mostrar as datas
# Configuramos o formato da data e a rotação para melhor legibilidade
ax.xaxis_date() #O método xaxis_date() é usado para dizer ao Matplotlib que as datas estão sendo usadas no eixo x
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)

# Adicionando título e rótulos para os eixos x e y
plt.title("Gráfico de Candlestick - PETR4.SA com matplotlib")
plt.xlabel("Data")
plt.ylabel("Preço")

# Adicionando uma grade para facilitar a visualização dos valores
plt.grid(True)

# Exibindo o gráfico
plt.show()

```

```python
# Criando subplots
'''
"Primeiro, criamos uma figura que conterá nossos gráficos usando make_subplots.
Isso nos permite ter múltiplos gráficos em uma única visualização.
Aqui, teremos dois subplots: um para o gráfico de candlestick e outro para o volume de transações."

'''
fig = make_subplots(rows=2, cols=1, shared_xaxes=True,
                    vertical_spacing=0.1,
                    subplot_titles=('Candlesticks', 'Volume Transacionado'),
                    row_width=[0.2, 0.7])

'''
"No gráfico de candlestick, cada candle representa um dia de negociação,
mostrando o preço de abertura, fechamento, máximo e mínimo. Vamos adicionar este gráfico à nossa figura."
'''
# Adicionando o gráfico de candlestick
fig.add_trace(go.Candlestick(x=df.index,
                             open=df['Abertura'],
                             high=df['Maximo'],
                             low=df['Minimo'],
                             close=df['Fechamento'],
                             name='Candlestick'),
                             row=1, col=1)

# Adicionando as médias móveis
# Adicionamos também médias móveis ao mesmo subplot para análise de tendências
fig.add_trace(go.Scatter(x=df.index,
                         y=df['MA7'],
                         mode='lines',
                         name='MA7 - Média Móvel 7 Dias'),
                         row=1, col=1)

fig.add_trace(go.Scatter(x=df.index,
                         y=df['MA14'],
                         mode='lines',
                         name='MA14 - Média Móvel 14 Dias'),
                         row=1, col=1)

# Adicionando o gráfico de barras para o volume
# Em seguida, criamos um gráfico de barras para o volume de transações, que nos dá uma ideia da atividade de negociação naquele dia
fig.add_trace(go.Bar(x=df.index,
                     y=df['Volume'],
                     name='Volume'),
                     row=2, col=1)

# Atualizando layout
#Finalmente, configuramos o layout da figura, ajustando títulos, formatos de eixo e outras configurações para tornar o gráfico claro e legível.
fig.update_layout(yaxis_title='Preço',
                  xaxis_rangeslider_visible=False,  # Desativa o range slider
                  width=1100, height=600)

# Mostrando o gráfico
fig.show()
```

Esse código parece ser uma chamada para gerar um gráfico de candlestick para os dados da ação da Apple (AAPL) usando a biblioteca `mplfinance` em Python. Vamos analisar linha por linha:

1. `mpf.plot`: Esta função é usada para plotar gráficos usando a biblioteca `mplfinance`, que é uma extensão da Matplotlib projetada especificamente para plotar gráficos financeiros.

2. `dados_apple.head(30)`: Esta é a parte dos dados que está sendo plotada. `dados_apple` parece ser um DataFrame contendo dados históricos de preços da Apple, e `.head(30)` está sendo usada para pegar as primeiras 30 linhas desse DataFrame.

3. `type='candle'`: Este argumento especifica o tipo de gráfico que será desenhado. No caso, é um gráfico de candlestick, que é comumente usado na análise técnica de ações e outros instrumentos financeiros.

4. `figsize=(16,8)`: Este argumento define o tamanho da figura do gráfico. Neste caso, a figura terá 16 polegadas de largura por 8 polegadas de altura.

5. `volume=True`: Este argumento especifica se o volume de negociação deve ser mostrado no gráfico. Aqui, está definido como verdadeiro, o que significa que o volume será mostrado.

6. `mav=(7,14)`: Este argumento especifica as médias móveis que serão plotadas no gráfico. Neste caso, duas médias móveis serão plotadas: uma de 7 períodos e outra de 14 períodos.

7. `style='brasil'`: Este argumento especifica o estilo do gráfico. 'brasil' é um dos estilos fornecidos pela `mplfinance`, que define cores e aparência adequadas para gráficos financeiros em um contexto brasileiro.

Juntando tudo, esta chamada de função irá plotar um gráfico de candlestick para os primeiros 30 dias de dados históricos da ação da Apple, mostrando também o volume de negociação e duas médias móveis (de 7 e 14 períodos), com um estilo específico para o Brasil.
